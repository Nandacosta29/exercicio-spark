```python
# 1. Instalar o PySpark
!pip install pyspark

# 2. Clonar o repositório com os dados
!git clone https://github.com/jbruno97/ecossistema-hadoop.git

# 3. Iniciar a Sessão Spark
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, sum, count

spark = SparkSession.builder \
    .appName("ExercicioHadoop") \
    .getOrCreate()

print("✅ Ambiente configurado e repositório clonado!")
```

```python
# Definir o caminho para a pasta clonada
path = "/content/ecossistema-hadoop/dados_colab/"

# Ler os arquivos CSV
df_vendas = spark.read.csv(path + "vendas.csv", header=True, inferSchema=True)
df_clientes = spark.read.csv(path + "clientes.csv", header=True, inferSchema=True)
df_produtos = spark.read.csv(path + "produtos.csv", header=True, inferSchema=True)

# Registrar tabelas para uso com Spark SQL
df_vendas.createOrReplaceTempView("vendas")
df_clientes.createOrReplaceTempView("clientes")
df_produtos.createOrReplaceTempView("produtos")

print("✅ DataFrames criados e Tabelas SQL registradas!")
```

```python
# Join entre vendas e produtos
df_vendas_produtos = df_vendas.join(df_produtos, 'produto_id')

# Agrupamento e soma
exercicio_1 = df_vendas_produtos.groupBy('categoria', 'nome_produto') \
    .agg(sum('preco_total').alias('faturamento')) \
    .orderBy(col('faturamento').desc())

print("Análise de Vendas por Categoria e Produto:")
exercicio_1.show(10)
```

```python
# 1. Instalação e Imports
!pip install pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, sum, count

# 2. Iniciar Sessão
spark = SparkSession.builder.appName("ExercicioCorrigido").getOrCreate()

# 3. Clonar Repositório e Ler Dados
!git clone https://github.com/jbruno97/ecossistema-hadoop.git
path = "/content/ecossistema-hadoop/dados_colab/"

df_vendas = spark.read.csv(path + "vendas.csv", header=True, inferSchema=True)
df_clientes = spark.read.csv(path + "clientes.csv", header=True, inferSchema=True)
df_produtos = spark.read.csv(path + "produtos.csv", header=True, inferSchema=True)

# 4. Registrar Tabelas para SQL
df_vendas.createOrReplaceTempView("vendas")
df_clientes.createOrReplaceTempView("clientes")
df_produtos.createOrReplaceTempView("produtos")

print("✅ Ambiente pronto e variáveis definidas!")
```

```python
df_vendas_prod = df_vendas.join(df_produtos, 'produto_id')
exercicio1_df = df_vendas_prod.groupBy('categoria', 'nome_produto') \
    .agg(sum('preco_total').alias('faturamento')) \
    .orderBy(col('faturamento').desc())
exercicio1_df.show(5)
```

```python
exercicio1_sql = spark.sql('''
    SELECT p.categoria, p.nome_produto, SUM(v.preco_total) as faturamento
    FROM vendas v
    JOIN produtos p ON v.produto_id = p.produto_id
    GROUP BY p.categoria, p.nome_produto
    ORDER BY faturamento DESC
''')
exercicio1_sql.show(5)
```

```python
df_vendas_cli = df_vendas.join(df_clientes, 'cliente_id')
exercicio2_df = df_vendas_cli.groupBy('nome') \
    .agg(sum('preco_total').alias('total_gasto')) \
    .orderBy(col('total_gasto').desc())
exercicio2_df.show(5)
```

```python
exercicio2_sql = spark.sql('''
    SELECT c.nome, SUM(v.preco_total) as total_gasto
    FROM vendas v
    JOIN clientes c ON v.cliente_id = c.cliente_id
    GROUP BY c.nome
    ORDER BY total_gasto DESC
    LIMIT 10
''')
exercicio2_sql.show(5)
```

```python
# Exemplo usando dataset de casas da Califórnia (nativo do Colab)
df_custom = spark.read.csv('/content/sample_data/california_housing_train.csv', header=True, inferSchema=True)

# Aplicação de conceito: Agrupar média de valor da casa por idade da casa
df_custom.groupBy('housing_median_age') \
    .agg(sum('median_house_value').alias('valor_total_por_idade')) \
    .orderBy('housing_median_age') \
    .show(5)
```
